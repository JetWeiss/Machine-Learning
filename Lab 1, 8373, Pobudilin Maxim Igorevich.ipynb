low: Самая низкая цена в конкретную дату (float)
close: Цена закрытия в конкретную дату (float)
volume: Количество купленного или проданного актива, отображаемого в базовой валюте (float)
marketCap: Общая стоимость всех добытых монет (float)
timestamp: Временная метка UTC рассматриваемого дня (datetime)
crypto_name: Название криптовалюты (string)
date: Преобразование timestamp в date (date)
Набор данных можно использовать для:
- Анализа временных рядов для различных криптовалют на основании даты
- Прогнозирования цен на различные криптовалюты
- Выполнения исследовательского анализа данных в наборе данных

Получение данных для анализа

```python
import pandas as pd
file_in = 'Cryptocurrency.csv'
events = pd.read_csv(file_in)
events.columns = ['id','open','high','low','close','volume','marketCap','timestamp','crypto_name','date']
events['timestamp'] = pd.to_datetime(events['timestamp'])
events_lite = events[events['crypto_name'] == 'Litecoin']
events_lite.head()
```

Определение среднего значения, ско для атрибутов: open, high, low, close, marketCap

```python
average_open = events_lite['open'].mean()
average_high = events_lite['high'].mean()
average_low = events_lite['low'].mean()
average_close = events_lite['close'].mean()
average_volume = events_lite['volume'].mean()
average_marketCap = events_lite['marketCap'].mean()
std_open = events_lite['open'].std()
std_high = events_lite['high'].std()
std_low = events_lite['low'].std()
std_close = events_lite['close'].std()
std_volume = events_lite['volume'].std()
std_marketCap = events_lite['marketCap'].std()
print('Для цены открытия:\n среднее значение - {} \n среднеквадратическое отклонение - {}'.format(average_open, std_open))
print('Для максимальной цены:\n среднее значение - {} \n среднеквадратическое отклонение - {}'.format(average_high, std_high))
print('Для минимальной цены:\n среднее значение - {} \n среднеквадратическое отклонение - {}'.format(average_low, std_low))
print('Для цены закрытия:\n среднее значение - {} \n среднеквадратическое отклонение - {}'.format(average_close, std_close))
print('Для количества проданного актива:\n среднее значение - {} \n среднеквадратическое отклонение - {}'.format(average_volume, std_volume))
print('Для Общей стоимости всех добытых монет:\n среднее значение - {} \n среднеквадратическое отклонение - {}'.format(average_marketCap, std_marketCap))
```

Построение гистрограмм распределения значений

```python
import matplotlib.pyplot as plt

plt.hist(events_lite['open'], edgecolor='black')
plt.title('Гистограмма распределения значений цены открытия')
plt.show()
plt.hist(events_lite['high'], edgecolor='black')
plt.title('Гистограмма распределения значений максимальной цены')
plt.show()
plt.hist(events_lite['low'], edgecolor='black')
plt.title('Гистограмма распределения значений минимальной цены')
plt.show()
plt.hist(events_lite['close'], edgecolor='black')
plt.title('Гистограмма распределения значений цены закрытия')
plt.show()
plt.hist(events_lite['volume'], edgecolor='black')
plt.title('Гистограмма распределения значений количества проданного актива')
plt.show()
plt.hist(events_lite['marketCap'], edgecolor='black')
plt.title('Гистограмма распределения значений общей стоимости монет')
plt.show()
```
